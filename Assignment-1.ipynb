```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

* Reading the dataset

```python
data_iris = pd.read_excel("iris.xls")
```

```python
data_iris
```

* Columns in the dataset


```python
data_iris.columns
```


* Calculating the mean of numeric dataset

```python
data_iris.mean(numeric_only= True)
```

* Null values present in the dataset

```python

data_iris.isnull().sum()
```

* Displays the distribution of different types of iris data

```python

plt.style.use("ggplot")
plt.figure(figsize=(10,4))
data_iris["Classification"].value_counts().plot.bar(color = ["lightcoral", "indianred","brown"])
plt.xticks(rotation=45)
plt.xlabel("Classification", labelpad = 20, fontsize = 12, color = "black")
plt.ylabel("Total count", labelpad = 20, fontsize = 12, color = "black")
plt.title("Classification of iris data", pad = 20, color = "darkslategray")

plt.show()
```

 * Distribution of each numerical data

```python

plt.figure(figsize=(18,10))

plt.subplot(2,2,1)
sns.histplot(data = data_iris, x = "SL", bins = 20, element="step", color="seagreen" )
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.subplot(2,2,2)
sns.histplot(data = data_iris, x = "SW", bins = 20, element="step", color="teal")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.subplot(2,2,3)
sns.histplot(data = data_iris, x = "PL", bins = 20, element="step", color="lightseagreen")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.subplot(2,2,4)
sns.histplot(data = data_iris, x = "PW", bins = 20, element="step", color = "darkslategray")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

* Shows how different numerical values are distributed among different classifications

```python

plt.figure(figsize=(18,7))

plt.subplot(1,2,1)
plt.bar(x = data_iris["Classification"],height=data_iris["SL"], width =0.3, label = "SL")
plt.bar(x = data_iris["Classification"],height=data_iris["SW"], align = "edge", width = 0.3, label = "SW")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Classification", labelpad = 20, fontsize = 18, color = "black")
plt.ylabel("Values", labelpad = 20, fontsize = 18, color = "black")
plt.legend(loc="upper center", shadow=True, fontsize = "large", facecolor = "oldlace")

plt.subplot(1,2,2)
plt.bar(x = data_iris["Classification"],height=data_iris["PL"], width =0.3, label = "PL")
plt.bar(x = data_iris["Classification"],height=data_iris["PW"], align = "edge", width = 0.3, label = "PW")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Classification", labelpad = 20, fontsize = 18, color = "black")
plt.ylabel("Values", labelpad = 20, fontsize = 18, color = "black")
plt.legend(loc="upper center", shadow=True, fontsize = "large", facecolor = "oldlace")

plt.show()
```

* Relation between each numeric value


```python

sns.heatmap(data = data_iris.corr(numeric_only = True), annot = True, cmap = "Blues")
```

* Displays how different classifications are scattered between the SL value and PL value

```python

sns.scatterplot(x="SL", y = "PL", data = data_iris, hue = "Classification")
```
