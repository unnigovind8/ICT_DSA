### 5. Check and handle outliers in at least 3 columns in the dataset

* for Age column

```python

plt.figure(figsize=(10,4))
sns.boxplot(data = titanic_data1, x = "Age")

plt.show()
```

```python
#sns.histplot(data = titanic_data1, x="Age", kde = True)
```

```python
data = titanic_data1["Age"]
# Calculating the IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

IQR
```

```python
threshold = 1.5

# Identify the outliers
outliers = (data < Q1 - threshold * IQR) | (data > Q3 + threshold * IQR)

titanic_data1[outliers]
```

```python
#removing outliers
titanic_no_outliers = titanic_data1[~outliers]
titanic_no_outliers.head(10)
```

```python
print(titanic_no_outliers.shape)
```

 * for fare column

```python

plt.figure(figsize=(10,4))
sns.boxplot(data = titanic_no_outliers, x = "Fare")

plt.show()
```

```python
data1 = titanic_no_outliers["Fare"]
#IQR for Fare column
Q1 = titanic_no_outliers["Fare"].quantile(0.25)
Q3 = titanic_no_outliers["Fare"].quantile(0.75)
IQR = Q3 - Q1

IQR
```

```python
threshold = 1.5

# Identifing the outliers
outliers1 = (data1 < Q1 - threshold * IQR) | (data1 > Q3 + threshold * IQR)

titanic_no_outliers[outliers1]
```

```python
#removing outliers from Fare column and adding remaining data with titanic_no_outliers

titanic_fare_no_outliers = titanic_no_outliers[~outliers1]

titanic_fare_no_outliers
```

 * for SibSp column

```python
sns.boxplot(data = titanic_fare_no_outliers, x = "SibSp")
```

```python
#calculating Z score
data2 = titanic_fare_no_outliers["SibSp"]

z_score = (data2 - data2.mean())/ data2.std()

z_score
```

```python
thresh = 2

outliers2 = abs(z_score) > thresh

titanic_fare_no_outliers[outliers2].head()
```

```python
titanic_fare_no_outliers[outliers2].shape
```

```python
# removing outliers and adding to new dataframe

titanic = titanic_fare_no_outliers[~outliers2]

titanic
```

### 6. Do min max scaling on the feature set (Take ‘Survived’ as target)

```python
#droping colums Name and ticket

titanic = titanic.drop(columns=["Name","Ticket"])
```

```python
titanic
```

```python
titanic.info()
```

```python
# Encoding Sex column using Label encoding

from sklearn.preprocessing import LabelEncoder

colum_to_encode = titanic["Sex"]

#initializing label encoder
label_encoder = LabelEncoder()

#fitting and transfroming values
encoded_value = label_encoder.fit_transform(colum_to_encode)


titanic["Sex"] = encoded_value
```

```python
titanic
```

```python
#Encodeing Embarked column using one_hot encoding

titanic["Embarked"].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder

colencode = titanic["Embarked"]

#converting to 2D array format
col_encode_2D = col_encode.values.reshape(-1,1)

onehot_encoder = OneHotEncoder()

encodedcol = onehot_encoder.fit_transform(col_encode_2D).toarray()

#creating a dataframe
encoded_df = pd.DataFrame(encodedcol, columns = onehot_encoder.get_feature_names_out(["Embarked"]))

encoded_df
```

```python
encoded_df.set_index(titanic.index, inplace = True)
```

```python
#combining the dataframes
titanic_encodeddf = pd.concat([titanic, encoded_df], axis =1)

titanic_encodeddf
```

```python
titanic_encodeddf.drop(columns="Embarked", inplace=True)
```

 * Normalizing using MinMax Scaling

```python
from sklearn.preprocessing import MinMaxScaler

#separating target variable from features
X = titanic_encodeddf.drop(columns = "Survived")
Y = titanic_encodeddf["Survived"]

scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)

scaled_df = pd.DataFrame(X_scaled, columns = X.columns)
scaled_df
```

```python
scaled_df.set_index(titanic_df.index, inplace=True)
scaled_df
```
